```python
#Histogram of Life_Expectancy Analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# visualization
import seaborn as sns
# Importing the dataset
dataset =pd.read_csv('/content/Life Expectancy Data.csv')

# Histogram
plt.hist(dataset['Life_Exp'], bins=20)
plt.xlabel('Life_Exp')
plt.ylabel('Frequency')
plt.title('Distribution of Life Expectancy')
plt.show()

```

```python
#plot() is used to plot the graph
dataset.plot()
```

```python
#finding the yearwise life expectancy analysis
dataset.groupby('Year')['Life_Exp'].count().plot(kind='pie',autopct='%1.1f%%', title='Year Wise Life Expectancy')
```

```python
# finding the top 10 countries with the highest life expectancy values
dataset.groupby('Country')['Life_Exp'].sum().sort_values(ascending=False).head(10).plot(kind='pie', autopct='%1.1f%%',title='top 10 countries with the highest sum of life expectancy values')

```

```python
# finding the top 10 countries with the highest population
dataset.groupby('Country')['Population'].sum().sort_values(ascending=False).head(10).plot(kind='pie', autopct='%1.1f%%',title='top 10 countries with the highest population')
```

```python
# Analysis of Year wise Adult Mortality rate
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dataset =pd.read_csv('/content/Life Expectancy Data.csv')
dataset.groupby('Year')['Adult Mortality'].count().plot(kind='pie',autopct='%1.1f%%' ,title='Year Wise Adult Mortality')
```
